```python
# 3. 데이터 불러오기
df = pd.read_csv('seoul.csv')
col_list = ['date','station_number','station','member','gender','age','use','momentum','carbon','distance(M)','time']
df.columns = col_list
df = df.replace(['70대이상','~10대','f','m'],['70대~','10대','F','M'])
df.head(200)
```

## 1. 이용건수 살펴보기

### 1) 대여소 별 이용건수 확인하기
 - 대여소별 이용건수 총합 구하기
 - 데이터 시각화하기

```python
# 대여소별 이용건수 총합 구하기
rent_use = df.groupby(df['station']).sum()['use']
rent_use = rent_use.reset_index()
rent_use
```

```python
# 데이터 시각화하기
X = rent_use['station']
y = rent_use['use']
plt.figure(figsize = (20,10))
plt.title('대여소별 이용건수')
plt.bar(X,y)
plt.show()
```

### 2) 연령대별 이용건수 시각화 해보기
 - 연령대별 이용건수 총합 더하기
 - 데이터 시각화하기

```python
age_use = df.groupby(df['age']).sum()['use']
age_use = age_use.reset_index()
print(age_use)
young = sum(age_use['use'][0:3])
old = sum(age_use['use'][3:8])
age_use_dict = {'age':['10~30대','그 외 연령층'],'user':[young,old]}
age_use_sum = pd.DataFrame(age_use_dict)
age_use_sum
```

```python
# 연령대별 사용비율 알아보기
wedgeprops = {'width' : 0.75, 'edgecolor' : 'w', 'linewidth' : 3}
colors = ['#009FE3', 'lightgrey']

plt.pie(age_use_sum['user'], 
        labels = age_use_sum['age'], 
        startangle = 90,
        autopct='%.1f%%',
        counterclock=False,
       wedgeprops = wedgeprops,
       colors = colors)
plt.title('연령별 따릉이 이용 비율')
plt.show()
```

```python
# 연령대별 이용건수 비교해보기
x = age_use['age']
y = age_use['use']
plt.bar(x,y,color = '#009FE3')
plt.title('연령별 따릉이 이용건수')
plt.show()
```

### 3) 성별 이용건수 시각화 해보기
 - 성별별 이용건수 총합구하기
 - 데이터 시각화하기

```python
# 응답 안한 성별 제거
gender_use = df[df['gender'] != 'NaN']
gender_use = gender_use.groupby(gender_use['gender']).sum()['use']
gender_use = gender_use.reset_index()
gender_use
```

```python
colors = ['lightgrey', '#009FE3']
explode = [0,0.05]

plt.pie(gender_use['use'], 
        labels = gender_use['gender'],
        autopct='%.1f%%',
       colors = colors,
       startangle = 90,
       explode = explode)
plt.title('성별별 따릉이 이용 비율')
plt.show()
```

### 4) 회원유형별 이용건수 시각화 해보기
 - 회원유형별 이용건수 총합구하기
 - 데이터 시각화하기

```python
member_use = df.groupby(df['member']).sum()['use']
member_use = member_use.reset_index()
member_use
```

```python
x = member_use['member']
y = member_use['use']
plt.bar(x,y,color = '#009FE3')
plt.title('회원유형별 이용건수')
plt.show()
```
